### Bitwise Operations

In Python, you can perform bitwise operations on integers using a set of built-in operators. These operations work at the binary level and are commonly used for tasks like setting, clearing, or toggling specific bits.



```python
# and 
a = 5      # Binary: 0101
b = 3      # Binary: 0011
result = a & b  # Binary: 0001 -> Decimal: 1
print(result)   # Output: 1
```

```python
# or
a = 5      # Binary: 0101
b = 3      # Binary: 0011
result = a | b  # Binary: 0111 -> Decimal: 7
print(result)   # Output: 7
```

```python
# xor
a = 5      # Binary: 0101
b = 3      # Binary: 0011
result = a ^ b  # Binary: 0110 -> Decimal: 6
print(result)   # Output: 6
```

```python
# not
a = 5      # Binary: 0101
result = ~a  # Binary: ...11111010 (2's complement) -> Decimal: -6
print(result)   # Output: -6
```

```python
# left shift
a = 5      # Binary: 0101
result = a << 1  # Binary: 1010 -> Decimal: 10
print(result)    # Output: 10
```

```python
# right shift
a = 5      # Binary: 0101
result = a >> 1  # Binary: 0010 -> Decimal: 2
print(result)    # Output: 2
```

**Combining Operations**

```python
a = 5      # Binary: 0101
b = 3      # Binary: 0011
result = (a & b) | (a ^ b)  # Perform AND, then XOR, then OR
print(result)  # Output: 7
```

```python
# Checking if a Number is Odd or Even:

num = 7
if num & 1:  # Check the least significant bit
    print("Odd")
else:
    print("Even")
```

```python
# Set a specific bit [Merging]

num = 5  # Binary: 0101
bit_position = 1
num |= (1 << bit_position)  # Set bit at position 1
print(num)  # Output: 7 (Binary: 0111)
```

```python
# Toggle a specific bit

num = 5  # Binary: 0101
bit_position = 0
num ^= (1 << bit_position)  # Toggle bit at position 0
print(num)  # Output: 4 (Binary: 0100)
```

```python
# Clear a specific bit

num = 5  # Binary: 0101
bit_position = 2
num &= ~(1 << bit_position)  # Clear bit at position 2
print(num)  # Output: 1 (Binary: 0001)
```
